```python
#dependent and independent data
Y = df.iloc[:,-1].values
X = df.iloc[:,:-1].values
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_new = sc.fit_transform(X)
```

```python
#test and train data
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X_new,Y,test_size=0.3,random_state=1)
```

```python
from sklearn.svm import SVC
sc = SVC()
```

```python
sc.get_params()
```

```python
sc.fit(X_train,Y_train)
```

```python
y_predict = sc.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,y_predict))
```

```python
#plot 
plt.scatter(df['meanfreq'],df['sd'])
plt.plot()
```

```python
#user defined function
from sklearn.model_selection import cross_val_score
#optimized value of C
accuracy_score=[]
for i in range(1,26):
    sc = SVC(C=i,kernel='rbf',random_state=1)
    score= cross_val_score(sc,X_train,Y_train,cv=5)
    accuracy_score.append(score.mean())
```

```python
plt.plot(range(1,26),accuracy_score)
plt.show()
```

```python
#user defined function
from sklearn.model_selection import cross_val_score
#optimized value of C
accuracy_sce=[]
for i in range(1,26):
    sc = SVC(C=i,kernel='rbf',random_state=1)
    score= cross_val_score(sc,X_test,Y_test,cv=5)
    accuracy_sce.append(score.mean())
```

```python
plt.plot(range(1,26),accuracy_sce)
plt.show()
```

```python
#user defined function
from sklearn.model_selection import cross_val_score
#optimized value of C
accuracy_score=[]
for i in list(np.arange(0.1,5,0.1)):
    sc = SVC(C=i,kernel='rbf',random_state=1)
    score= cross_val_score(sc,X_train,Y_train,cv=5)
    accuracy_score.append(score.mean())
```

```python
plt.plot(list(np.arange(0.1,5,0.1)),accuracy_score)
plt.show()
```

```python
#user defined function
from sklearn.model_selection import cross_val_score
#optimized value of C
gama = [0.001,0.01,0.1,1,10,100]
g_score= []
for i in gama:
    sc = SVC(C=1,kernel='rbf',random_state=1,gamma=i)
    score= cross_val_score(sc,X_train,Y_train,cv=5)
    g_score.append(score.mean())
```

```python
plt.plot(gama,g_score)
plt.scatter(gama,g_score)
for i in range(len(g_score)):
    plt.text(gama[i],g_score[i],(gama[i],g_score[i]))
plt.grid()
plt.show()
```
